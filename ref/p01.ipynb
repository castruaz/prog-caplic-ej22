- mayor a 90 y menor que 180 Obtuso

## p33 - Operaciones sobre listas
- Inicializar de forma predeterminada un arreglo
- Efectuar las operaciones basicas

````python
# inicializar lista y efectuar operaciones basicas

numeros = [10,20,30,40,60,70,10,20,99]

print(f"El arreglo:{numeros}")

print(f"Longitud de la lista: {len(numeros)}")
print(f"Primer elemento     : {numeros[0]}")
print(f"Ultimo elemento     : {numeros[-1]}")

print("\nRecorrido de los elementos:")
for n in numeros:
  print(n, end=" ")

print("\n\nRecorrido de elemntos por :")
for i in range(len(numeros)):
  print(numeros[i], end=" ")

print("\n\nCopiar y eliminar elementos:")
otros = numeros.copy()
print(f"La copia es    : {otros}")
print(f"Elimina primero: {otros.pop(3)} -> {otros}")
otros.remove(10)
print(f"Elimna varios  : {otros}")
otros.clear()
print(f"Elimina todos  : {otros}")
otros = numeros.copy()

print("\nAgregar, buscar y contar elementos:")
print(f"Al empezar     : {otros}")
otros.extend([200,300,400])
print(f"Extender       : {otros}")
otros.append(500)
print(f"Agregar final  : {otros}")
otros.insert(10,333)
print(f"Insertar  333  : {otros}")
print(f"Buscar 99      : pos: {otros.index(99)}")
print(f"Contar 10s     : {otros.count(10)}")
otros = numeros.copy()

print("\nInvertir y ordenar")
print(f"Al empezar     : {otros}")
otros.reverse()
print(f"Invertido      : {otros}")
otros.sort()
print(f"Ordenado       : {otros}")
````

## p36 - Listas con aletatorios
- Genera 2 listas de 10 numeros aleatorios
- Eleva todos sus numeros al cuadrado
- Suma ambas listas y genera una nueva

````python
# listas con aleatorios
import random
l1 = []
l2 = []
l3 = []

print("Generando aleatorios ...")
for n in range(10):
  l1.append(random.randint(1,100))
  l2.append(random.randint(1,100))
  
print("\nListas originales:")
print(f"Lista 1 : {l1}")
print(f"Lista 2 : {l2}")

for n in range(10):
  l1[n]=pow(l1[n],2)
  l2[n]=pow(l2[n],2)
  l3.append(l1[n] + l2[n])

print("\nListas procesadas:")
print(f"Lista 1 : {l1}")
print(f"Lista 2 : {l2}")
print(f"Lista 3 : {l3}")
````

## p49 - suma digitos
Dado un número entero, sumar sus dígitos


````python
def sumadigitos(n):
  suma=0
  while n!=0:
    dig = n % 10
    suma = suma + dig
    n = n // 10
  return suma


n = int(input("Dame un numero y sumaré sus dígitos :? "))
res = sumadigitos(n)
print(f"La suma de los dígitos de {n} es {res}")
````

## p50 - Factorial
Dado un número calcular su factorial

````python
def factorial(numero):
    f = 1
    for n in range(1,numero+1):
      f = f * n
    return f

num = int(input("Dame un número ? "))
print(f"El factorial de {num} es {factorial(num)}" )
````

## p51 - Calificacion con letra
Dada una calificacion entera, imprimir la letra correspondiente de acuerdo a lo siguiente
```
90-100 A Excelente
80-89  B Bien 
70-79  C Suficiente
60-69  D Deficiente
0-59  F  Reprobado
```


````python
def califletra(cal):
  if cal>=90 and cal<=100:
    return "A", "Excelente"
  elif cal>=80 and cal<=89:
    return "B", "Bien"
  elif cal>=70 and cal<=79:
    return "C", "Suficiente"
  elif cal>=60 and cal<=69:
    return "D", "Deficiente"
  elif cal>=0 and cal<=59:
    return "F", "Reporbado"

cal = int(input("Dame una calificacion entre 1 y 100 ?"))
l , m = califletra(cal)
print(f"Tu calificacion de {cal} corresponde a {l} y esta {m}")
````


## p52 - Pasando arreglo a funciones
Calculo de estadisticas:
- mayor
- menor
- media
- varianza
- desviación estándar

````python
import math

def mostrar(nums):
  for i in range(0,len(nums)):
    print(f"{nums[i]} ", end="")

def mayormenor(nums):
  may=men=nums[0]
  for i in range(1,len(nums)):
    if nums[i]>may: may=nums[i]
    if nums[i]<men: men=nums[i]
  return may, men

def media(nums):
  s=0
  for i in range(0,len(nums)):
    s=s+nums[i]
  return s/len(nums)

def varianza(nums, rmedia):
  var = 0
  for i in range(0,len(nums)):
    var = var + ( nums[i] - rmedia  ) ** 2
  return var / (len(nums) -1)

def leerdatos():
  datos=[]
  while True:
    d=int(input("número: "))
    if d==-1:
      break;
    datos.append(d)
  return datos

nums = [2,5,5,6,8,8,9,2]
#nums=leerdatos()

print("Numeros: ")
mostrar(nums) 
print("\nEstadisticas con los elementos del arreglo:")
rmayor, rmenor = mayormenor(nums)
rmedia = media(nums)
rvaria = varianza(nums,rmedia)
rdesv  = math.sqrt(rvaria)

print(f"Total de datos  : {len(nums)}")
print(f"El mayor es     : {rmayor}")
print(f"El menor es     : {rmenor}")
print(f"La media es     : {rmedia}")
print(f"La varianza   es: {rvaria}")
print(f"La desviacion es: {rdesv}")
````
