```python
import pandas as pd
edades = {"Juan":40, "Pedro":20, "Maria":36, "Francisco":20, "Luis":34}
datos = pd.Series(edades)
datos.name = 'Edades'
# edad de juan
print(datos[0], '\n')
# Los 3 primeros (indices 0,1,2)
print(datos[:3], '\n')
#los 3 ultimos (indices 2,3,4 )
print(datos[-3:], '\n')
# elementos consecutivos (indice 1,2,3)
print(datos[1:4], '\n')
# acceder a elementos en orden inverso'
print(datos[::-1], '\n')
 
```

```python
import pandas as pd
edades = {"Juan":40, "Pedro":20, "Maria":36, "Francisco":20, "Luis":34}
datos = pd.Series(edades)
datos.name = 'Edades'

# edad Pedro 
print( datos['Pedro'], '\n' ) 

# edades Juan y Maria
print( datos[['Juan','Maria']], '\n' )

# edades desde Juan hasta Mara
print( datos['Juan':'Maria'] ,'\n')
```

```python
import pandas as pd
edades = {"Juan":40, "Pedro":20, "Maria":36, "Francisco":20, "Luis":34}
datos = pd.Series(edades)
datos.name = 'Edades'

# estadistica descriptiva
print( datos.describe(), '\n')
 
print( datos.count() )          # Cuenta
print( datos.mean() )           # Media
print( datos.std() )            # Desv estandar
print( datos.min() )            # Minimo
print( datos.quantile(q=0.25) ) # Quartil 25%
print( datos.quantile(q=0.50) ) # Quartil 50%
print( datos.quantile(q=0.75) ) # Quartil 75%
print( datos.max() )            # Maximo
 
```

```python
import pandas as pd
edades = {"Juan":40, "Pedro":20, "Maria":36, "Francisco":20, "Luis":34}
datos = pd.Series(edades)
datos.name = 'Edades'

 
print( datos.sum(), '\n') # la suma de todas las edades
print( datos[0:3].sum(), '\n') # suma edades de 0 a 3
print( datos[['Juan','Pedro']].sum(), '\n') # suma edades Juan y Pedro
print( datos['Pedro':'Francisco'].sum(), '\n') # suma edades Juan y Pedro
print( datos[[0,2,4]].mean(), '\n') # media de indices 0,2,4
print( datos[2:4].std(), '\n') # desvacion de indices 2,3
print( datos+3, '\n') # suma 3 a todas las edades
print( datos*2, '\n') # multiplica x 2 todas las edades


```

## DataFrame

### DataFrame en base a varias series

```python
# Crear varias Series
import pandas as pd
nombre   = pd.Series(["Juan", "Pedro", "Luis", "Maria", "Genaro", "Israel", "Fatima", "Angelica", "Ancelmo"])
edad     = pd.Series([33, 35, 35, 30, 25, 34, 42, 22, 55])
peso     = pd.Series([45, 65.5, 30.5, 80, 45, 25, 60, 45, 90])
estatura = pd.Series([1.22, 1.70, 1.80, 1.66, 1.50, 1.90, 1.33, 1.75, 1.80])
# Crear DataFrame en base a las Series
df = pd.DataFrame(data={"Nombre":nombre, "Edad":edad, "Peso":peso, "Estatura":estatura})
df.set_index('Nombre', inplace=True)
```

## Anális 1 - Club que tiene los mejores jugadores Real Madrid vs FC Barcelona

```python
rm = fifa[ fifa['club_name']=='Real Madrid' ]  # dataframe con jugadores del: Real Madrid
br = fifa[ fifa['club_name']=='FC Barcelona' ] # dataframe con jugadores del: FC Barcelona
rm.shape, br.shape
```

```python
# etadísticas básicas por club en base a overall (calificación general de los jugadores)
print(rm['overall'].describe()) 
print(br['overall'].describe())
```

```python
# Diagrama de caja y bigotes: Real Madrid
plt.figure()
plt.boxplot(rm['overall'])
plt.title('Real Madrid')
plt.ylabel('Overall')
plt.grid(True)

# Diagrama de caja y bigotes: FC Barcelona
plt.figure()
plt.boxplot(br['overall'])
plt.title('FC Barcelona')
plt.ylabel('Overall')
plt.grid(True)
```

## Anális 1 - Club que tiene los mejores jugadores más valorados ? Real Madrid vs FC Barcelona

```python
# etadísticas básicas por club (valor de los jugadores en euros)
print(rm['value_eur'].describe()) 
print(br['value_eur'].describe())
```

```python
# Histograma de ambos clubs en un mismo escenario
plt.figure()
plt.hist(rm['value_eur'])
plt.xlabel('Valor en 10x8 de millones de Euros')
plt.ylabel('Frecuencia')
plt.title('Real Madrid')
plt.figure()
plt.hist(br['value_eur'])
plt.xlabel('Valor en 10x8 de millones de Euros')
plt.ylabel('Frecuencia')
plt.title('FC Barcelona')
```

```python
# jugadores del Real Madrid con valor mayor a 50 millones de euros
br[ br['value_eur'] > 50000000 ]
```

```python
# jugadores del FC Barcelona con valor mayor a 50 millones de euros
br[ br['value_eur'] > 50000000 ]
```
